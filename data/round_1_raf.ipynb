```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
days = [-2, -1, 0]
prices_df = [pd.read_csv(f"capsule/prices_round_1_day_{day}.csv", sep=";") for day in days]
trades_df = [pd.read_csv(f"capsule/trades_round_1_day_{day}.csv", sep=";") for day in days]
print(prices_df[0].head())
print(trades_df[0].head())
```

```python
price_cols = list(filter(lambda x: x not in ["day", "product", "profit_and_loss"], prices_df[0].columns))
osmium_prices = pd.concat([
    df[df["product"] == "ASH_COATED_OSMIUM"].assign(timestamp=df["timestamp"]+d*1_000_000)[price_cols]
    for d, df in enumerate(prices_df)
]).set_index("timestamp").sort_index()
print(osmium_prices)
```

```python
trade_cols = ["timestamp", "symbol", "price", "quantity"]
osmium_trades = pd.concat([
    df[df["symbol"] == "ASH_COATED_OSMIUM"].assign(timestamp=df["timestamp"]+d*1_000_000)[trade_cols]
    for d, df in enumerate(trades_df)
]).set_index("timestamp").sort_index()
print(osmium_trades)
```
